```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
p = np.linspace(0.01, 0.74, 74)
```

```python
plt.plot(p, p)
plt.plot(p, -np.log(1 - p))
plt.plot(p, 0.7 * ((1 - p) ** (-1 / 0.7) - 1))
plt.plot(p, -0.75 * np.log(1 - p * 4 / 3))
plt.xlim(0, 0.74)
plt.show()
```
